### **Compounds can be classified according to solubility values (LogS); Compounds with 0 and higher solubility value are highly soluble, those in the range of 0 to ‚àí2 are soluble, those in the range of ‚àí2 to ‚àí4 are slightly soluble and insoluble if less than ‚àí4. lets find out how many of them are highly soluble, slightly soluble and insoluble**

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Load data from CSV file
data_path = '/content/output.csv'
df = pd.read_csv(data_path)

# Binarize actual and predicted solubility based on the threshold
threshold = -4
df['y_binary'] = df['y'].apply(lambda x: 1 if x >= threshold else 0)
df['Solubility_binary'] = df['Solubility'].apply(lambda x: 1 if x >= threshold else 0)

# Calculate confusion matrix
conf_matrix = confusion_matrix(df['y_binary'], df['Solubility_binary'])

# Plot confusion matrix heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Calculate accuracy, precision, recall, and F1 score
accuracy = accuracy_score(df['y_binary'], df['Solubility_binary'])
precision = precision_score(df['y_binary'], df['Solubility_binary'])
recall = recall_score(df['y_binary'], df['Solubility_binary'])
f1 = f1_score(df['y_binary'], df['Solubility_binary'])

# Print evaluation metrics
print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
